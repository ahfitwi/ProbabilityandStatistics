# <font color='green'>4. Probability</font>

### <font color='green'>4.1 What is Probability</font>

- A value between 0 and 1 that a certain even will occur.
- The probability that a fair coin will come up head is 0.5
- $P(E_{head})\ =\ 0.5$, 50\% of chance of occurrence
- Trial: the act of flipping a coin
- Each trial is independent of all others
- Independent trials have no memory of the history of trials
- ***Experiment***: Each trial of flipping a coin  
- ***Simple Event***: Each mutually exclusive outcome. Rolling a siz-sided die: $E1=1, E2=2, E3=3, E4=4, E5=5, E6=6$
- ***Sample Space***: The sum of every possible simple event. $S\ =\ \{E1, E2, E3, E4, E5, E6\}$
- The probability a fair-die will roll 6 is $P(E=6)\ =\ \frac{1}{6}$

***Example 1***: A company made a total of 50 trumpet valves. It is determined that one the valves is defective. If three valves go into one trumpet, what is the probability that a trumpet has a defective valve?

***Solution:***

$$P(valve = Defective)\ =\ \frac{1}{50}\ =\ 0.02 $$

$$P(Trumpet = Defective)\ =\ 3*P(valve = defective) $$
$$P(Trumpet = Defective)\ =\ 3* 0.02\ =\ 0.06\ =\ 6\% $$


### <font color='green'>4.2 Permutation</font>

- A permutation of a set of objects is an arrangement of the objects in a certain order

- n factorial = n!

***Permutation without repetition:***

Take a subset of items in a permutation: the number of permutations of a set of n objects taken r at a time is given by the following formula:

$$nPr = \frac{n!}{(n-r)!}$$

***Example 2***: A website requires a 4 character password characters can either be lowercase letters or the digits 0-9. You may not repeat a letter or number.

***Solution:*** $$\frac{(26+10)!}{(26+10-4)!}\ =\ 36*35*34*33\ =\ 1, 413, 720$$

***Permutation with repetition:*** $$n^r$$
    

***Example 3***: How many 4 digit license plates can you make using the numbers 0 to 9 while allowing repetition.

**Solution**: $10^4$

### <font color='green'>4.3 Combination</font>

- Unordered arrangemens of objects are called combinations
- A group of people selected for a team are the same group, no matter the order.

***Combination without repetition***
- The number of combinations of a set of n objects taken r at a time without a repetition is given by:

$$nCr\ =\ \frac{n!}{r!(n-r)!}$$

***Example 4***: How many 3-letter combinations can be made from the letters ABCDE?

**Solution**: $$5C3\ =\ \frac{5!}{3!(5-3)!}\ =\ \frac{5*4}{2}\ =10$$

***Example 5***: For a study, 4 people are chosen at random from a group of 10 people. How many ways can this be done?

**Solution**: $$10C4\ =\ \frac{10!}{4!(10-4)!}\ =\ \frac{10*9*8*7}{4*3*2}\ =210$$

```python
from math import factorial as f
f(10)/(f(4)*f(6))
```

***Combination with repetition***
$$n+r-1Cr\ =\ \frac{(n+r-1)!}{r!(n-1)!}$$

***Example 6***: For a pizza, 4 ingredients are chosen at random from a possible of 10 ingredients. How many different pizza topping combinations are there allowing repetition?

***Solution***: $$n+r-1Cr\ =\ \frac{(10+4-1)!}{4!(10-1)!}\ =\ \frac{13*12*11*10}{4*3*2}\ =\ 715$$

***Summary***:
- Permutation without repetition: $nPr = \frac{n!}{(n-r)!}$
- Permutation with repetition: $nPr = n^r$
- Combination without repetition: $nCr = \frac{n!}{r!(n-r)!}$
- Combination without repetition: $n+r-1Cr = \frac{n+r-1}{r!(n-1)!}$

### <font color='green'>4.4 Intersections, Unions, and Complements</font>

- ***Intersection***: it describes the sample space where two events both occur. $$P(A\cap B)\ =\ \frac{A\cap B}{Sample Space}$$

- ***Union***: it describes two events if A or B occurs. $$P(A\cup B)\ =\ \frac{A\cup B}{Sample Space}$$

$$P(A\cup B)\ =\ P(A)+P(B)-P(A\cap B)$$

- ***Complement of an event***: considers everything outside of the event. $$P(\bar{A})\ =\ 1-P(A)$$

### <font color='green'>4.5 Independent and Dependent Events</font>

 An ***Independent*** series of events occur when the outcome of one event has no effect on the outcome of another.
- Filipping a fair coin twice: the chance of getting heads on the second toss is independent of the result of the first toss.
- The probability of seeing two heads with two flips a fair coin is:
$$P(H_1H_2)\ =\ P(H_1)xP(H_2)$$


A ***dependent*** event occurs when the outcome of a first event does affect the probability of a second event. Recall that when the outcome of a prior event affects the probability of a current event, the current event is said to be a dependent event.

***Example 7***: A common example is to draw colored marbles from a bag without replacement. From a container containing 5 balls (2 blue and 3 red), the probability of drawing a second red marble given that the first was red is written as :
$$P(R_2|R_1)\ =\ \frac{2}{4}$$
$$P(R_1\cap R_2)\ =\ P(R_1)P(R_2|R_1)\ =\ \frac{3}{5}* \frac{2}{4}=0.3$$

### <font color='green'>4.6 Conditional Probability</font>

It is the idea that we want to know the probability of event A, given that event B has occurred, is conditional probability. This is written as $P(A|B)$

$$P(A|B)\ =\ \frac{P(A\cap B)}{P(B)}$$

***Example 8***: Given that a project is completed on time B, what is the probability that it is under budget A? On Time = 48, Intersection =16, and under budget = 62.

***Solution***: $$P(A|B)\ =\ \frac{P(A\cap B)}{P(B)}\ =\ \frac{16}{48}\ =\ \frac{1}{3}$$

***Mutually Exclusive Events***: when two events cannot both happen, they are said to be mutually exclusive. $A\cap B\ =\ 0$

### <font color='green'>4.7 Addition and Multiplication Rules</font>

***Addition Rule***
$$P(A\cup B)\ =\ P(A)+P(B)-P(A\cap B)$$

***Example 9***: A standard deck of 52 cards is shuffled. How would you determine the probability that the top two cards are both clubs?

Using the ***Addition Rule***:
$$P(A\cup B)\ =\ P(A)+P(B)-P(A\cap B)$$
    
$$ 13/52 + 13/51 $$

***Multiplication Rule***

From the section on dependent events we saw that the probability of A and B is: 
$$P(A\cap B)\ =\ P(A)*P(B|A)$$

***Example 10***: Given a standard deck of 52 cards, what is the probability of drawing a aces?

Using the ***A Multiplication Rule***:
$$P(A\cap B\cap C\cap D)\ =\ P(A)*P(B|A)*P(C|AB)*P(D|ABC)$$
$$P(A\cap B\cap C\cap D)\ =\ \frac{4}{52}*\frac{3}{51}*\frac{2}{50}*\frac{1}{49}\ =\ \frac{1}{270, 725}$$   


### <font color='green'>4.8 Bayes Theorem</font>

If $P(B) > 0$:
$$P(A|B)\ =\ \frac{P(A\cap B)}{P(B)}$$

If $P(A) > 0$:
$$P(B|A)\ =\ \frac{P(B\cap A)}{P(A)}$$

Bayes Theorem is used to determine the probability of a aparameter, given a certain event. The general formula where $P(A) > 0$ and $P(B) > 0$ is:
$$P(A|B)\ =\ \frac{P(B|A)*P(A)}{P(B)}$$

***Example 10***: A company learns that 1 out of 500 of their products are defective, or 0.2%. The company buys a diagnostic tool that correctly identifies a defective part 99\% of the time. If a part is diagnosed as defective, what is the probability that it really is defective.

***Solutions***: $P(A|B)\ =\ \frac{P(B|A)*P(A)}{P(B)}$

- p(A|B) = probability of being defective if testing positive = ?
- p(B|A) = probability of testing positive if defective = 0.99
- p(A) = probability of being defective = 0.002
- p(B) = probability of testing positive = $P(TP) + P(FP)$
    - **True Positive**: product is defective and gets classified as defective
    - **False Positive**: product is not defective, but still gets classified as defective.
    - $P(B) = P(B|A)*P(A)+P(B|\bar{A}.P(\bar{A})$
    - $P(B) = 0.99*0.002+0.01*0.998\ =\ 0.01196 $
$$P(A|B)\ =\ \frac{0.99*0.002}{0.01196}=0.165$$

A positive test has a 16.5\% chance of correctly identifying a defective part.

***Extended problem***: what if we perform a second test, and that also comes up positive?
$$P(A|B)\ =\ \frac{0.99*0.165}{0.99*0.165+0.01*0.835}=0.951$$

***Example 11***: In a company of 100 employees, 10 are accountants. For some reason most of the accountants, 8 out of 10, wear glasses, while only 20% of all employees wear glasses. If you see an employee wearing glasses, what is the probability that they are an accountant?

***Solution***:
- P(A|B): Probability of being accountant if wearing glas = ?
- P(A): probability of employees who are accountant
- P(B|A): Probability of an accountant who wears a glass 
- P(B): Probability of employees that wear glass

   - $P(A|B) = ?$
   - $P(A) = 0.1$
   - $P(B|A) = 0.8$
   - $P(B) = 0.2$
        

$$P(A|B)\ =\ \frac{P(B|A)*P(A)}{P(B)}$$
$$P(A|B)\ =\ \frac{0.8*0.1}{0.2}\ =\ 0.4$$

### <font color='green'>5.3 Bernoulli Distribution</font>

- A ***Bernoulli Trial*** is a random experiment in which there are only two possible outcomes, success or failure

- The Bernoulli distribution is a discrete distribution having two possible outcomes labeled by $x=1$ and $x=0$ in which $x= 1$ represents a "success" that occurs with a probability $p$ and $x=0$ represents a "failure" that occures with probability $q=1-p$, where $0\le p\le 1$. Whence, the range space is $R_x = \{0, 1\}$.
- It therefore has a probability density function of

\begin{equation}
  P(x)=\begin{cases}
    1-p, & \text{for $x = 0$}.\\    
    p, & \text{for $x = 1$}.
  \end{cases}
\end{equation}

$$ P(x)\ =\ p^x(1-p)^{1-x}$$

$$E(x)\ =\ 1*p +0*(1-p)\ =\ p$$
$$V(x)\ =\ 1*p -p^2\ =\ p(1-p)$$

- The performance of a fixed number of trials with fixed probability of success on each trial is known as a Bernoulli trial. It is used to model the probability of an outcome having desired class or characteristics, for instance
    - A computer system is up or down (on or off)
    - A packet in a computer network reaches or doesn’t reach its destination
    - A bit in a given packet is affected by noise and arrives in erroneous state.

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
from random import randint

k1 = [randint(0,1) for i in range(50)]
p1=0.5

p = [bernoulli.pmf(k, p1) for k in k1]
plt.bar(k1,p)
plt.grid()
plt.show()
```

### <font color='green'>5.4 Geometric Distribution</font>

- The geometric distribution represents the number of failures before you get a success in a series of Bernoulli trials.
- This discrete probability distribution is represented by the probability density function (Probability mass function):

$$f(x)\ =\ (1-p)^{x-1}p$$

- $k\ge 1$
- For example, you ask people outside a polling station who they voted for until you find someone that voted for the independent candidate in a local election. 
- The geometric distribution would represent the number of people who you had to poll before you found someone who voted independent. 
- You would need to get a certain number of failures before you got your first success. 
- If you had to ask 3 people, then $X=3$; if you had to ask 4 people, then $X=4$ and so on. In other words, there would be $X-1$ failures before you get your success. 
- If $X=n$, it means you succeeded on the $n^{th}$ try and failed for n-1 tries. 
- The probability of failing on your first try is $1-p$.

- ***For example***, if p = 0.2 then your probability of success is 0.2 and your probability of failure is 1 – 0.2 = 0.8. 

- Independence (i.e. that the outcome of one trial does not affect the next) means that you can multiply the probabilities together. So the probability of failing on your second try is $(1-p)(1-p)$ and your probability of failing on the $n^{th}-1$ tries is $(1-p)n-1$. 
- If you succeeded on your $4^{th}$ try, $n = 4$, $n – 1 = 3$, so the probability of failing up to that point is $(1-p)(1-p)(1-p) = (1-p)^3$.

***Example***: If your probability of success is 0.2, what is the probability you meet an independent voter on your third try?

***Solution***:
$$f(x)\ =\ (1-p)^{x-1}p$$
$$P(x=3)\ =\ (1-0.2)^{3-1}0.2\ =\ 0.128$$

### <font color='green'>5.6 Poisson Distribution</font>

- A binomial distribution considers the number of successes out of n trials
- A Poisson Distribution considers the number of successes per unit of time* over the course of many units* or any other continuous unit, e.g. distance
- Extensively used in queuing model
- Calculation of the Poisson probability mass function starts with a mean expected value

$$E(X) = \mu$$

- This is then assigned to “lambda” (arrival rate)

$$\lambda\ =\ \frac{\#\ occurrences}{Interval}\ =\ \mu$$

- Number of successes per continuous intervals
-  $\lambda\$ = average number of cars per hour

- The equation becomes
$$P(x)\ =\ \frac{\lambda^x e^{-\lambda}}{x!}$$

where $e\ =\ Euler's\ number\ =\ 2.71828 ...$

- $R_x: \{0, 1,2,3, ..., \infty \}$
- $E(x)\ =\ \lambda t$
- $V(x)\ =\ \lambda t$
- $\lambda$ is the expected arrival rate

***Example***: A warehouse typically receives 8 deliveries between 4 and 5pm on Friday. What is the probability that only 4 deliveries will arrive between
4 and 5pm this Friday?

***Solution:***
$x\ =\ 4,\ \lambda\ =\ 8$ (It means on average)
$$P(x)\ =\ \frac{\lambda^x e^{-\lambda}}{x!}=\frac{8^4*2.71828^{-8}}{4!}\ =\ 0.0572$$

    - Change of variable: $ t\ =\ \frac{(x- \mu)}{\sigma} $
$$ F_X(x)\ =\ \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\frac{x-\mu}{\sigma}}e^{\frac{-t^2}{2\sigma}}dt\ =\ \phi(\frac{x-\mu}{\sigma})$$
    - If $\mu\ =\ 0\ and\ \sigma\ =\ 1$:
$$\phi(x)\ =\ \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{x}e^{\frac{-t^2}{2}}dt\ $$

### <font color='green'>5.9 Relations between the Distributions</font>

#### Relation between Bernoulli and Binomial Distribution

1. Bernoulli Distribution is a special case of Binomial Distribution with a single trial.

2. There are only two possible outcomes of a Bernoulli and Binomial distribution, namely success and failure.

3. Both Bernoulli and Binomial Distributions have independent trails

$$As\ n\ \rightarrow\ 1\ :\ Bernoulli$$

$$P(x:n, p)\ =\ \binom{n}{x}p^x(1-p)^{n-x}\  \rightarrow\ p^x(1-p)^{1-x}$$

#### Relation between Poisson and Binomial Distribution
Poisson Distribution is a limiting case of binomial distribution under the following conditions:
- The number of trials is indefinitely large or n → ∞.
- The probability of success for each trial is same and indefinitely small or p →0.
- np = λ, is finite

Essentially, we may approximate the binomial probability distribution with the probability of Poisson distribution whenever $n$ is large, and $p$ is small.
$$ As\ n\ \rightarrow \infty\, \ p\ \rightarrow 0,\ and\ \alpha\ =\ np$$
$$p(x\ =\ k)\ =\ \binom{n}{k} p^k (1-p)^{n-k}\ \rightarrow\ p(x\ =\ k)\ =\ \frac{e^{-\alpha} \alpha^k}{k!}$$

#### Relation between Normal and Binomial Distribution:
Normal distribution is another limiting form of binomial distribution under the following conditions:
- The number of trials is indefinitely large, $n\ \rightarrow \infty$.
- Both p and q are not indefinitely small.

The normal distribution can be used as an approximation to the binomial distribution, under certain circumstances, namely:
   - If X ~ B(n, p) and if n is large and/or p is close to ½, then X is approximately N(np, npq), (where q = 1 - p).
   - N(np, npq) = N(mean, variance)
   
   $$p(x\ =\ k)\ =\ \binom{n}{k} p^k (1-p)^{n-k}\ \rightarrow\ \frac{1}{(npq)\sqrt{2\pi}}e^{\frac{-(x-np)^2}{2(npq)^2}}$$

#### Relation between  Normal and Poisson Distribution:
- The normal distribution is also a limiting case of Poisson distribution with the parameter $\lambda\ \rightarrow \infty$.
- If X ∼Poisson (λ)  ⇒ X ≈N ($\mu\ =\ \lambda,\ \sigma\ =\ \sqrt{\lambda}$), for $\lambda\ >\ 20$, and the approximation improves as (the rate) $\lambda$ increases.

$$P(x)\ =\ \frac{\lambda^x e^{-\lambda}}{x!}\ \rightarrow\ \frac{1}{\sqrt{2\pi \lambda}}e^{\frac{-(x-\lambda)^2}{2\lambda}}$$


#### Relation between Exponential and Poisson Distribution:

If the times between random events follow exponential distribution with rate $\lambda$, then the total number of events in a time period of length $t$ follows the Poisson distribution with parameter $\lambda t$.

$$f(x)\ =\ \lambda e^{-\lambda x}\ \rightarrow\ \frac{(\lambda t)^x e^{-(\lambda t)}}{x!}$$
   - $\ x\ge 0\ and\ \lambda >0.\ \lambda$  is the failure rate of a device at any time t, given that it has survived upto $t$.
